```python
import sys
sys.path.append("C:\\Users\\jernelson\\Development\\jpn-app-dev\\Lib\\site-packages")
import redis
from pymongo import Connection
analytics = redis.StrictRedis()
datastore = Connection(port=27018)
print("Total keys in analytics={}\n\tIII Items={}".format(analytics.dbsize(), analytics.get('iii-item')))
print("Top ten books")
top_ten = []
for r in analytics.zrevrange('item-checkouts', 0, 10, withscores=True):
    item_number = analytics.hget(r[0], 'number')
    marc = datastore.marc.bibliographic.find_one({"fields.945.subfields.y": item_number},
                                                 {"fields.100", "fields.245"})
    authors, title = [], str()
    for field in marc.get('fields'):
        if len(field) > 0:
            tag = field.keys()[0]
            value = ''.join([subfield.get('a','') for subfield in field[tag].get('subfields')])
            if tag.startswith('100'):
                authors.append(value)
            if tag.startswith('245'):
                title = value
    top_ten.append({"title": title, "count": r[1]})
    print("\t{} by {} {} Checkouts. Redis key {} datastore ID {}".format(title, 
                                                                       ' '.join(authors), 
                                                                       r[1],
                                                                       r[0],
                                                                       marc.get("_id")))
```

```python
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(1)
last_year = datetime.datetime.utcnow().year
years = [year for year in xrange(last_year-11, last_year)]
checkouts = [analytics.bitcount('{}:checkouts'.format(year)) for year in years]
plt.plot(years, checkouts, 'ro-')
plt.ylabel('Number of checkouts')
plt.xlabel('Year')
plt.title("Patron checkouts per Year")
```

```python
plt.figure(2)
plt.title("10 Most Checked Out Books")
plt.plot([x.get('count') for x in top_ten])
plt.xlabel([x.get('title') for x in top_ten])

```
