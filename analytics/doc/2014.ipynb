```python
import sys
sys.path.append("C:\\Users\\jernelson\\Development\\jpn-app-dev\\Lib\\site-packages")
import helpers
import redis
from pymongo import Connection
analytics = redis.StrictRedis()
datastore = Connection(port=27018)
print("Total keys in analytics={}\n\tIII Items={}".format(analytics.dbsize(), analytics.get('iii-item')))
print("Top 100 items")
top_ten = []
for r in analytics.zrevrange('item-checkouts', 0, 100, withscores=True):
    item_number = analytics.hget(r[0], 'number')
    marc = datastore.marc.bibliographic.find_one({"fields.945.subfields.y": item_number},
                                                 {"fields.100", "fields.245"})
    authors, title = [], str()
    for field in marc.get('fields'):
        if len(field) > 0:
            tag = field.keys()[0]
            value = ''.join([subfield.get('a','') for subfield in field[tag].get('subfields')])
            if tag.startswith('100'):
                authors.append(value)
            if tag.startswith('245'):
                title = value
    top_ten.append({"title": title, "count": r[1]})
    print(u"\t{} by {} {} Checkouts. Redis key {} datastore ID {}".format(title, 
                                                                       ' '.join(authors), 
                                                                       r[1],
                                                                       r[0],
                                                                       marc.get("_id")))
    
```

```python
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(1)
last_year = datetime.datetime.utcnow().year
years = [year for year in xrange(last_year-15, last_year)]
checkouts = [analytics.bitcount('{}:checkouts'.format(year)) for year in years]
plt.plot(years, checkouts, 'ro-')
plt.ylabel('Number of checkouts')
plt.xlabel('Year')
plt.title("Patron Checkouts per Year")
plt.show()
for i,year in enumerate(years):
    print("{}: {}".format(year, checkouts[i]))
```

```python
from collections import OrderedDict
months = OrderedDict({0:'January', 1:'February', 2:'March'})
plt.figure(3)
plt.title("Circulation Activity January-March 2014")
circ_stats_q1 = [analytics.bitcount('2014-0{}:checkouts'.format(month+1)) for month in months.keys()]
plt.plot(circ_stats_q1)
plt.xticks(months.keys(), months.values())
plt.xlabel("Months")
plt.ylabel("Number of checkouts")
plt.show()

```

```python
import helpers

# January status
print("Total checkouts for January 2014 is {}".format(analytics.bitcount("2014-01:checkouts")))
jan_stats = helpers.get_month_stats('checkouts', 2014, 1, analytics)
plt.plot([x[1] for x in jan_stats])
plt.title("Checkouts for January 2014")
plt.xlabel("Day of the week")
plt.ylabel("Number of checkouts")
plt.show()
print("Total checkouts for February 2014 is {}".format(analytics.bitcount("2014-02:checkouts")))
feb_stats = helpers.get_month_stats('checkouts', 2014, 2, analytics)
plt.plot([x[1] for x in feb_stats])
plt.title("Checkouts for February 2014")
plt.xlabel("Day of the week")
plt.ylabel("Number of checkouts")

plt.show()
print("Total checkouts for March 2014 is {}".format(analytics.bitcount("2014-03:checkouts")))
mar_stats = helpers.get_month_stats('checkouts', 2014, 3, analytics)
plt.plot([x[1] for x in mar_stats])
plt.title("Checkouts for March 2014")
plt.xlabel("Day of the week")
plt.ylabel("Number of checkouts")
plt.show()
first_quarter = [x[1] for x in jan_stats]
first_quarter.extend([x[1] for x in feb_stats])
first_quarter.extend([x[1] for x in mar_stats])
plt.title("Daily Checkouts Jan-Mar 2014")
plt.plot(first_quarter)
plt.show()
```
